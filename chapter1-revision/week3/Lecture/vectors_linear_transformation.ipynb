```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
vector_a = np.array([3,1]).reshape((2,1))
# print(vector_a.shape, vector_a)
vector_b = np.array([1,2]).reshape((2,1))

vector_matrix = np.hstack([vector_a,vector_b])
linear_transformation_one = np.dot(vector_matrix, np.array([0,0]))
                                   
print(linear_transformation_one.shape,linear_transformation_one)

linear_transformation_two = np.dot(vector_matrix, np.array([1,1]))

print(linear_transformation_two.shape,linear_transformation_two)
                                   
origin = np.array([[0, 0],[0, 0]])
origin_1d = np.array([[0, 0]])

# plt.quiver(*origin, vector_a, vector_b, color=['r','b'], scale=10)
plt.quiver(*origin, vector_a, vector_b, linear_transformation_one, color=['r', 'b', 'g'], scale=10)
plt.quiver(*origin_1d, linear_transformation_two, color=['g'], scale=10)





plt.show()
```
