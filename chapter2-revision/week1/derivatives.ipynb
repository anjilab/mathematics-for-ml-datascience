```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Non differentiable at pointed end.
x = np.array([1,2, -1])
y1 = np.abs(x)
plt.plot(x,y1, color='red')
# plt.plot(x,y2, color='green')
```

```python
# Discontinuity
x = np.array([1,2, -1])
y1 = np.array([2,2,2])
y2 =  x + 1
plt.plot(x,y1, color='red')
plt.plot(x,y2, color='red')
plt.xlim(-2,7)
# plt.ylim)

# np.piecewise(x, [x < 0, x >= 0], [lambda x: -x, lambda x: x


```

```python
# discontinuity
def f(x):
    if(x == 2): return 0
    else: return 1

x = np.arange(0., 5., 0.2)

y = []
for i in range(len(x)):
   y.append(f(x[i]))

print(x)
print(y)

plt.plot(x,y,c='red', ls='', ms=5, marker='.')
ax = plt.gca()
ax.set_ylim([-1, 2])

plt.show()

```

```python
# VERITICAL TANGENT, SEE AT X = 0, STRAIGHT through y-axis, vertical tangent, slope = 0
x = np.linspace(0,20,1000)
y1 =  np.power(x,1/3)
plt.plot(x,y1, color='red')
plt.ylim(-2,5)
# plt.xlim(-2,2)
```
