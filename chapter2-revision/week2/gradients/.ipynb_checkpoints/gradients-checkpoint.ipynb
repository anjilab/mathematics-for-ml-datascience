```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d
```

```python
# Functions with one variable. y is in terms of x only i.e y = f(x), derivative will be 2x + 2 which is simple line
x = np.linspace(-10, 10, 1000)

# calculate the y value for each element of the x vector
y = x**2 + 2*x + 2 
y_dash = 2*x + 2 # this is the tangent line.



fig, ax = plt.subplots()
ax.plot(x, y)
ax.plot(x, y_dash)
```

```python
# Functions with 2 variable, f(x,y) = x^2 + y^2
# in such functions tangents won't be line instead it will be plane. 
# so what plane is it ?
# takinng 2 cases, cut spaces to plane.
# take x = 2, we get 2d parabola and find its tangent
# take y = 4, we get another 2d parabola, & find its tangent
# the two tangents intersect each other, its unique plane.
# treating y as constant, the given function will be of 1 variables i.e x
# treating x as constant, the given function will be of 1 variables i.e y
# for function f(x,y), take partial derivatives w.r.t each variable.


def function(x, y):
    # return np.sin(np.sqrt(x ** 2 + y ** 2))
    return np.power(x,2) + np.power(y,2)

def derivative_function_w_r_t_x(x):
    # return np.sin(np.sqrt(x ** 2 + y ** 2))
    return 2*x


def derivative_function_w_r_t_y(y):
    # return np.sin(np.sqrt(x ** 2 + y ** 2))
    return 2*y

def plane(x,y):
    return 2*x + 2*y
 
 
x = np.linspace(-10, 10, 40)
y = np.linspace(-10, 10, 40)
 
X, Y = np.meshgrid(x, y)
Z = function(X, Y)
tangent_z_parallel_y = plane(X,Y)
tangent_z_parallel_x = plane(Y,X)

 
fig = plt.figure(figsize=(10, 8))
ax = plt.axes(projection='3d')
 
ax.plot_surface(X, Y, Z, cmap='cool', alpha=0.8)

# st line
# ax.plot(x,derivative_function_w_r_t_x(x))
# ax.plot(y,derivative_function_w_r_t_y(y))
ax.plot_surface(X, Y, tangent_z_parallel_y, alpha=0.5)

# ax.plot(x,function(x,4))
# ax.plot(x,function(2,y))

 
ax.set_title('3D Contour Plot of function(x, y) =\
               x^2 + y^2', fontsize=14)
ax.set_xlabel('x', fontsize=12)
ax.set_ylabel('y', fontsize=12)
ax.set_zlabel('z', fontsize=12)
 
plt.show()
```

```python
# We now know that, f(x,y), slicing in 2 ways, treating y as constant, we get f/x & treating x as constant f/dy.
# These partial derivative forms a vector called gradient(vector containing partial derivative of given function).
# This gradient will outlines about the slope of two lines forming tangent plane.
```

Derivative of a function with single variable is useful for finding max-min by setting it to 0. used for optmization.
Similarly, gradient of a function with 2 variables is useful for finding max-min.
For one function to find minimum point, slope will be equal to 0. To solve we take derivative & set it to 0 to get the minimum value.        In case of function with 2 variables, we find the slope of tangent lines in tangent plane are given by the partial derivatives setting to 0. So, if there is case of 12 variables, we need to finding 12 lines with slope to 0 to find minimum.
