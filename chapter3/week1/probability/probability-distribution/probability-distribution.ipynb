```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# When coin is flipped 3 times, X=NO OF HEADS, P(X) ,HERE X = RANDOM VARIABLE and its possible values are {0,1,2,3}, P(X=1) is an event. 
```

```python
# Discrete distribution
x = np.array([0,1,2,3])
p_x = np.array([1/8,3/8,3/8,1/8])

plt.bar(x,p_x)
```

```python
# Cumulative distribution
x = np.array([0,1,2,3])
p_x = np.array([1/8,3/8,3/8,1/8])
c_x = np.array([1/8, (1/8+3/8) , (1/8+3/8+3/8) , 1])
plt.bar(x,c_x)
```

```python
mu, sigma = 2, 0.1 # mean and standard deviation
s = np.random.normal(mu, sigma, 10000)
count, bins, ignored = plt.hist(s, 30, density=True)
plt.plot(bins, 1/(sigma * np.sqrt(2 * np.pi)) *
               np.exp( - (bins - mu)**2 / (2 * sigma**2) ),
         linewidth=2, color='r')
plt.show()
```
