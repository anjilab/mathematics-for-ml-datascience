## we can see that in model 3 it overfits, perfectly fits the data, but model 2 nearly fits the data. Hence to avoid such scenario, we introduce penality in each model. penalty = L2 regularization term where, sum of squares of variable.
In model 1 = 4*x + 3, L2= 4**2 = 16. After calculating penalty, we can know the loss by calculating difference in squares of actual * predicted. here in example we are supposing loss. After we find the loss, we add the net loss. whichever loss is small is the model for the given data.
