# Importance & ways to visualize data correctly.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = np.array([40,20,30,80,75,60,55,46,34,43,22,67,56,45])
data_prob = np.array([1/16,1/16, 1/16,1/16,1/16,1/16,1/16,1/16,1/16,1/16,1/16,2/16,2/16,1/16])

print(np.median(data))
```

```python
print("sorted data",data.sort())
print(data)
median_point = len(data)/2 
self_calculated_md = (data[int(median_point)] +  data[int(median_point)-1])/2
print(self_calculated_md)
```

## 1/4 = Q1(25%) Quartile 1 and 3/4 = Q4 (75%) = Quartile 3

### BOX PLOT / WHISKER PLOT, 
We need three quartiles, Q1, Q2, Q3.

```python
Q1_index = (len(data) + 1 ) / 4
Q1 = (data[int(Q1_index)] +  data[int(Q1_index)-1])/2
Q2_index = (len(data) + 1 ) / 2
Q2 = (data[int(Q2_index)] +  data[int(Q2_index)-1])/2
Q3_index = 3*(len(data) + 1 ) / 4
Q3 = (data[int(Q3_index)] +  data[int(Q3_index)-1])/2
print(Q1,Q2,Q3)


inter_quartile_range = Q3-Q1 # 50% OF YOUR DATA LIES WITHIN THIS RANGE
print(np.max(data), np.min(data))
upper_whisker_horizontal_point = np.min([Q3 + 1.5 * inter_quartile_range,np.max(data)])
lower_whisker_horizontal_point = np.max([Q1 - 1.5 * inter_quartile_range,np.min(data)]) 

print(upper_whisker_horizontal_point,lower_whisker_horizontal_point)
```

## Box plot helps us to analyze skewness.
If you observe below boxplot for the given data, we can see that, data is not so skewed, if the orange line was below towards Q1 then we could say skewed towards Q2 as Q3-Q2 will be larger in such case scenario. If skewed to Q1, the Q1-Q2 will be larger than that of Q3-Q2.

## density estimation
For continuous RV, we can use proability distribution function to describe the data. Histogram can be helpful to analyze such distribution as it follows the property that area under curve is 1, is always positive. But if we draw the histogram we can peaks, since continuous should be smooth. To do show we use kernel density estimation. First, draw obs along x-axis, secondly, put mountains i.e gaussian in each obs, you can use any but here gaussian. Multiply everything by 1/n and sum of the curves. for large data set, this kernel density estimation, maps the pdf well.

## ASSESSING NORMALITY OF DISTRIBUTION:

First: Standardize the data (x-/)
Second: Compute quantiles
Third: Compare to gaussian quantiles.

Our sample data should align or nearly align to the theoretical quantiles if they are gaussian in nature.
