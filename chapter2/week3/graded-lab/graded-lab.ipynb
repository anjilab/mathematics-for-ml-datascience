```python
sample_size = 200
samples, labels = make_blobs(n_samples=sample_size, 
                             centers=([2.5, 3], [6.7, 7.9], [2.1, 7.9], [7.4, 2.8]), 
                             cluster_std=1.1,
                             random_state=0)
# print(samples,labels)
print(samples.shape)
print(labels.shape)
print("Before",labels)
# There are 0,1,2,3 total of 4 labels observed in the data. We converted 0 and 1 to 1 and 2 & 3 to 0 class category.
labels[(labels == 0) | (labels == 1)] = 1
labels[(labels == 2) | (labels == 3)] = 0
print(labels)
X = np.transpose(samples)
Y = labels.reshape((1,sample_size))
print(X.shape, Y.shape)
plt.scatter(X[0, :], X[1, :], c=Y, cmap=colors.ListedColormap(['blue', 'red']));
```

```python
def sigmoid(z):
    return 1/(1+np.power(np.e,-z))

print("sigmoid(-2) = " + str(sigmoid(-2)))
print("sigmoid(0) = " + str(sigmoid(0)))
print("sigmoid(3.5) = " + str(sigmoid(3.5)))

test.test_sigmoid(sigmoid)
```

```python
def nn_two_layers(features_X,actual_Y):
    print(actual_Y.shape)
    print(features_X.shape)
    number_of_layers = 2
    n_x = features_X.shape[0]
    n_y = actual_Y.shape[0]
    return (n_x,number_of_layers,n_y)

print('Network with two layers:', nn_two_layers(X,Y))
test.test_layer_sizes(nn_two_layers)
    
```
