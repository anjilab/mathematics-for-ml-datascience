```python
sample_size = 2000
samples, labels = make_blobs(n_samples=sample_size, 
                             centers=([2.5, 3], [6.7, 7.9], [2.1, 7.9], [7.4, 2.8]), 
                             cluster_std=1.1,
                             random_state=0)
# print(samples,labels)
print(samples.shape)
print(labels.shape)
print("Before",labels)
# There are 0,1,2,3 total of 4 labels observed in the data. We converted 0 and 1 to 1 and 2 & 3 to 0 class category.
labels[(labels == 0) | (labels == 1)] = 1
labels[(labels == 2) | (labels == 3)] = 0
print(labels)
X = np.transpose(samples)
Y = labels.reshape((1,sample_size))
print(X.shape, Y.shape)
plt.scatter(X[0, :], X[1, :], c=Y, cmap=colors.ListedColormap(['blue', 'red']));
```

```python
def sigmoid(z):
    return 1/(1+np.power(np.e,-z))

print("sigmoid(-2) = " + str(sigmoid(-2)))
print("sigmoid(0) = " + str(sigmoid(0)))
print("sigmoid(3.5) = " + str(sigmoid(3.5)))

test.test_sigmoid(sigmoid)
```

```python
def nn_two_layers(features_X,actual_Y):
    print(actual_Y.shape)
    print(features_X.shape)
    number_of_layers = 2
    n_x = features_X.shape[0]
    n_y = actual_Y.shape[0]
    return (n_x,number_of_layers,n_y)

print('Network with two layers:', nn_two_layers(X,Y))
test.test_layer_sizes(nn_two_layers)
    
```

```python
def initial_parameters(n_x,n_h,n_y):
    w1 = np.random.rand(n_h,n_x) *0.01 # Weight from input layer to hidden layer
    b1 = np.zeros((n_h,1))# bias vector from input layer to hidden layer
    w2 = np.random.randn(n_y,n_h) # weight from hidden to output layer
    b2 = np.zeros((n_y,1))

    assert (w1.shape == (n_h, n_x))
    assert (b1.shape == (n_h, 1))
    assert (w2.shape == (n_y, n_h))
    assert (b2.shape == (n_y, 1))
    
    parameters = {"W1": w1,
                  "b1": b1,
                  "W2": w2,
                  "b2": b2}
    
    return parameters
    
    
    
```

```python
layer_sizes_value = nn_two_layers(X,Y)
print(layer_sizes_value)

parameters = initial_parameters(layer_sizes_value[0], layer_sizes_value[1], layer_sizes_value[2])

test.test_initialize_parameters(initial_parameters)
```

```python

def forward_propagation(X, parameters):
    """
    Argument:
    X -- input data of size (n_x, m)
    parameters -- python dictionary containing your parameters (output of initialization function)
    
    Returns:
    A2 -- the sigmoid output of the second activation
    cache -- python dictionary containing Z1, A1, Z2, A2 
    (that simplifies the calculations in the back propagation step)
    """
    # Retrieve each parameter from the dictionary "parameters".
    ### START CODE HERE ### (~ 4 lines of code)
    W1 = parameters["W1"]
    b1 = parameters["b1"]
    W2 = parameters["W2"]
    b2 = parameters["b2"]
    ### END CODE HERE ###
    
    # Implement forward propagation to calculate A2.
    ### START CODE HERE ### (~ 4 lines of code)
    Z1 = np.dot(W1, X) + b1
    A1 = sigmoid(Z1)
    Z2 = np.dot(W2,A1) + b2
    A2 = sigmoid(Z2)
    ### END CODE HERE ###
    
    assert(A2.shape == (layer_sizes_value[2], X.shape[1]))

    each_layer_output = {"Z1": Z1,
             "A1": A1,
             "Z2": Z2,
             "A2": A2}
    
    return A2, each_layer_output
    
# def forward_propagation(X, parameters):
#     print("Forward propagation")
#     w1 = parameters['W1']
#     b1 = parameters['b1']
#     w2 = parameters['W2']
#     b2 = parameters['b2']
#     print(w1.shape,X.shape)
#     z1 = np.dot(w1,X) + b1
#     a1 = sigmoid(z1)
#     z2 = np.dot(w2,a1) + b2
#     a2 = sigmoid(z2)

#     assert(a2.shape == (layer_sizes_value[2], X.shape[1]))
#     each_layer_output = {'A1': a1, 'Z1': z1, 'Z2': z2, 'A2':a2 }
#     return a2 , each_layer_output

a2,each_layer_output = forward_propagation(X, parameters)

test.test_forward_propagation(forward_propagation)

    
    
```

```python
def cost_function(A2,Y):
    print(A2,Y)
    number_of_dataset = Y.shape[1]
    logloss = np.dot(-1*Y, np.log(A2).T)-np.dot((1-Y),np.log(1-A2).T)
    cost = 1/number_of_dataset * logloss

    cost = cost[0][0]
    assert(isinstance(cost, float))
    
    return float(cost)
    
print("cost = " + str(cost_function(a2, Y)))

test.test_compute_cost(cost_function, a2)
    
    
```

```python
# Computing partial derivates: backward propagation, updating new parameters based on these values

# def backward_propagation(parameters,each_layer_output,X,Y):
#     number_of_dataset = Y.shape[1]
#     print("Backward propagation")
    
#     w1 = parameters["W1"]
#     w2 = parameters["W2"]
#     A2 = each_layer_output["A2"]
#     A1 = each_layer_output["A1"]
    
#     difference_layer_output_exact_op = A2 - Y
#     gradient_w_r_t_w2 = (1/number_of_dataset) * np.dot(difference_layer_output_exact_op, A1.T)
#     print(gradient_w_r_t_w2)
#     # below code once see carefully
#     gradient_w_r_t_b2 = 1/number_of_dataset * np.sum(difference_layer_output_exact_op, axis = 1, keepdims = True)
#     differences_w1 = np.dot(w2.T, difference_layer_output_exact_op) * A1 * (1-A1)
#     gradient_w_r_t_w1 = (1/number_of_dataset)* np.dot(differences_w1, X.T)
#     gradient_w_r_t_b1 = 1/number_of_dataset * np.sum(differences_w1, axis = 1, keepdims = True)
#     grads = {"dW1": gradient_w_r_t_w1,
#              "db1": gradient_w_r_t_b1,
#              "dW2": gradient_w_r_t_w2,
#              "db2": gradient_w_r_t_b2}
    
#     return grads
# gradients = backward_propagation(parameters,each_layer_output,X,Y)
    
    
    
def backward_propagation(parameters, cache, X, Y):
    """
    Implements the backward propagation, calculating gradients
    
    Arguments:
    parameters -- python dictionary containing our parameters 
    cache -- python dictionary containing Z1, A1, Z2, A2
    X -- input data of shape (n_x, number of examples)
    Y -- "true" labels vector of shape (n_y, number of examples)
    
    Returns:
    grads -- python dictionary containing gradients with respect to different parameters
    """
    m = X.shape[1]
    
    # First, retrieve W from the dictionary "parameters".
    W1 = parameters["W1"]
    W2 = parameters["W2"]
    
    # Retrieve also A1 and A2 from dictionary "cache".
    A1 = cache["A1"]
    A2 = cache["A2"]
    
    # Backward propagation: calculate partial derivatives denoted as dW1, db1, dW2, db2 for simplicity. 
    dZ2 = A2 - Y
    dW2 = 1/m * np.dot(dZ2, A1.T)
    db2 = 1/m * np.sum(dZ2, axis = 1, keepdims = True)
    dZ1 = np.dot(W2.T, dZ2) * A1 * (1 - A1)
    dW1 = 1/m * np.dot(dZ1, X.T)
    db1 = 1/m * np.sum(dZ1, axis = 1, keepdims = True)
    
    grads = {"dW1": dW1,
             "db1": db1,
             "dW2": dW2,
             "db2": db2}
    
    return grads

grads = backward_propagation(parameters, each_layer_output, X, Y)

print("dW1 = " + str(grads["dW1"]))
print("db1 = " + str(grads["db1"]))
print("dW2 = " + str(grads["dW2"]))
print("db2 = " + str(grads["db2"]))    
    
    
    



    
```

```python
def update_parameters(parameters, gradients, learning_rate=1.2):
    print("Updating parameters")
    w1 = parameters["W1"]
    w2 = parameters["W2"]
    b1 = parameters["W1"]
    b2 = parameters["b2"]
    
    w1 = w1 - learning_rate * gradients["dW1"]
    b1 = b1 - learning_rate * gradients["db1"]
    w2 = w2 - learning_rate * gradients["dW2"]
    b2 = b2 - learning_rate * gradients["db2"] 
    
    parameters = {"W1": w1,
                  "b1": b1,
                  "W2": w2,
                  "b2": b2}
    return parameters
    
    

print(update_parameters(parameters, gradients, 1.2))
test.test_update_parameters(update_parameters)
    
```

```python
# def nn_model(X,Y,n_h, num_iterations=10, learning_rate =1.2, print_cost= False):
#     n_x = nn_two_layers(X,Y)[0]
#     n_y = nn_two_layers(X,Y)[2]
    
#     parameters = initial_parameters(n_x, n_h, n_y)
#     for i in range(num_iterations):
#         a2, each_layer_output =  forward_propagation(X,parameters)
#         cost = cost_function(a2,Y)
#         grads = backward_propagation(parameters,each_layer_output,X.T,Y)
#         parameters = update_parameters(parameters,gradients,1.2)
#         if print_cost:
#             print("Cost is:",cost)
    
#     return parameters


def nn_model(X, Y, n_h, num_iterations=10, learning_rate=1.2, print_cost=False):
    """
    Arguments:
    X -- dataset of shape (n_x, number of examples)
    Y -- labels of shape (n_y, number of examples)
    num_iterations -- number of iterations in the loop
    learning_rate -- learning rate parameter for gradient descent
    print_cost -- if True, print the cost every iteration
    
    Returns:
    parameters -- parameters learnt by the model. They can then be used to predict.
    """
    
    n_x = nn_two_layers(X, Y)[0]
    n_y = nn_two_layers(X, Y)[2]
    
    # Initialize parameters.
    ### START CODE HERE ### (~ 1 line of code)
    parameters = initial_parameters(2,2,1)
    ### END CODE HERE ###
    
    # Loop.
    for i in range(0, num_iterations):
         
        ### START CODE HERE ### (~ 4 lines of code)
        # Forward propagation. Inputs: "X, parameters". Outputs: "A2, cache".
        A2, each_layer_output = forward_propagation(X,parameters)
        
        # Cost function. Inputs: "A2, Y". Outputs: "cost".
        cost = cost_function(A2,Y)
        
        # Backpropagation. Inputs: "parameters, cache, X, Y". Outputs: "grads".
        grads = backward_propagation(parameters, each_layer_output, X, Y)
        
        # Gradient descent parameter update. Inputs: "parameters, grads, learning_rate". Outputs: "parameters".
        parameters = update_parameters(parameters, grads, learning_rate)
        ### END CODE HERE ###
        
        # Print the cost every iteration.
        if print_cost:
            print ("Cost after iteration %i: %f" %(i, cost))

    return parameters


parameters = nn_model(X, Y, n_h=2, num_iterations=3000, learning_rate=1.2, print_cost=True)
print("W1 = " + str(parameters["W1"]))
print("b1 = " + str(parameters["b1"]))
print("W2 = " + str(parameters["W2"]))
print("b2 = " + str(parameters["b2"]))

W1 = parameters["W1"]
b1 = parameters["b1"]
W2 = parameters["W2"]
b2 = parameters["b2"]
            
        
        
test.test_nn_model(nn_model)  
        
        
```

```python
def plot_decision_boundary(predict, parameters, X, Y):
    # Define bounds of the domain.
    min1, max1 = X[0, :].min()-1, X[0, :].max()+1
    min2, max2 = X[1, :].min()-1, X[1, :].max()+1
    # Define the x and y scale.
    x1grid = np.arange(min1, max1, 0.1)
    x2grid = np.arange(min2, max2, 0.1)
    # Create all of the lines and rows of the grid.
    xx, yy = np.meshgrid(x1grid, x2grid)
    # Flatten each grid to a vector.
    r1, r2 = xx.flatten(), yy.flatten()
    r1, r2 = r1.reshape((1, len(r1))), r2.reshape((1, len(r2)))
    # Vertical stack vectors to create x1,x2 input for the model.
    grid = np.vstack((r1,r2))
    # Make predictions for the grid.
    predictions = predict(grid, parameters)
    # Reshape the predictions back into a grid.
    zz = predictions.reshape(xx.shape)
    # Plot the grid of x, y and z values as a surface.
    plt.contourf(xx, yy, zz, cmap=plt.cm.Spectral.reversed())
    plt.scatter(X[0, :], X[1, :], c=Y, cmap=colors.ListedColormap(['blue', 'red']));

# Plot the decision boundary.
plot_decision_boundary(predict, parameters, X, Y)
plt.title("Decision Boundary for hidden layer size " + str(layer_sizes_value[1]))
```
